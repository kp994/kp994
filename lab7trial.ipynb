```python
import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split
```

```python

from scipy import stats
import numpy as np


df = pd.read_csv('SolarPredictionfinal.csv')

column_to_clean = 'Radiation'

z_scores = stats.zscore(df[column_to_clean])

threshold = 3

outlier_indices = (np.abs(z_scores) > threshold)

df_no_outliers = df[~outlier_indices]

df_no_outliers.reset_index(drop=True, inplace=True)

# Display the cleaned dataset
print(df_no_outliers)
```

```python
features = df_no_outliers[['Humidity','Temperature']]
X = features.values  # Feature matrix

# Assuming 'Radiation' is your regression target
y_regression = df_no_outliers['Radiation'].values 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Calculate the mean of the column "Radiation"
threshold = df_no_outliers['Radiation'].mean()

# Transform Labels based on the mean
y_binary = np.where(df_no_outliers['Radiation'] > threshold, 1, 0)

# Split Data
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)

# Train Binary Classification Model
classifier = LogisticRegression()  # You can replace this with another classifier
classifier.fit(X_train, y_train)

# Evaluate the Model
y_pred = classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f'Threshold (mean of Radiation): {threshold}')
print(f'Accuracy: {accuracy}')
print(f'Classification Report:\n{report}')


```

```python
# Predict class labels for the test set
predicted_labels = clf.predict(X_test)

# Analyze the output values
for i, predicted_label in enumerate(predicted_labels):
    actual_label = y_test[i]  # Actual class label for the i-th test sample
    print(f"Predicted: {predicted_label}, Actual: {actual_label}")

# Calculate accuracy manually
correct_predictions = 0
total_samples = len(y_test)

for i in range(total_samples):
    if predicted_labels[i] == y_test[i]:
        correct_predictions += 1

accuracy = correct_predictions / total_samples
print(f"Manual Accuracy: {accuracy * 100:.2f}%")
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score

# Define a list of kernel functions to experiment with
kernel_functions = ['linear', 'poly', 'rbf', 'sigmoid']

# Create and evaluate SVM models for each kernel function
for kernel in kernel_functions:
    print(f"Kernel Function: {kernel}")
    
    # Create an instance of the SVM classifier with the specified kernel
    clf = svm.SVC(kernel=kernel)
    
    # Train the SVM on the training data
    clf.fit(X_train, y_train)
    
    # Predict class labels for the test set
    predicted_labels = clf.predict(X_test)
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, predicted_labels)
    
    print(f"Accuracy: {accuracy * 100:.2f}%\n")
```
